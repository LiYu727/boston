```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import keras
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.optimizers import SGD, Adam
from keras.datasets import boston_housing
```

```python
# Load the Boston housing dataset, this will need an active internet connection
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
```

```python
#Explore the data structure using basic python commands
print("Type of the Dataset:",type(y_train))
print("Shape of training data :",x_train.shape)
print("Shape of training labels :",y_train.shape)
print("Shape of testing data :",type(x_test))
print("Shape of testing labels :",y_test.shape)
```

```python
#Extract the last 100 rows from the training data to create thevalidation datasets.
x_val = x_train[300:,]
y_val = y_train[300:,]
```

```python
print("Shape of validation data :",x_val.shape)
print("Shape of  validation data :",y_val.shape)
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```python
#Define the model architecture
# Define the model
basic_model = Sequential()
basic_model.add(Dense(units=13, input_dim=13, kernel_initializer='normal',activation='relu'))
basic_model.add(Dense(units=6, kernel_initializer='normal',activation='relu'))
basic_model.add(Dense(units=1, kernel_initializer='normal', activation='linear'))
```

```python
#optimizer=Adam(lr=0.000001)
```

```python
# Compile the model
basic_model.compile(optimizer="adam", loss='mean_squared_error', metrics=['mean_absolute_percentage_error'])
```

- We have created a simple two-hidden-layer model for the regression use case. 
- We have chosen MAPE as the metric. 
- Generally, this is not the best metric to choose for studying model performance, but its advantage
  is simplicity in terms of comprehending the results. 
  It gives a simple percentage value for the error, say 10% error.

```python
# Train the model with validation split
#alidation_split=0.2
history = basic_model.fit(x_train, y_train, epochs=3, batch_size=32, validation_data=(x_val,y_val))
```

```python
import matplotlib.pyplot as plt

# Assuming 'history' is the variable that holds the output from the model.fit() method
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)

plt.figure(figsize=(10, 6))
plt.plot(epochs, loss, 'bo-', label='Training loss')
plt.plot(epochs, val_loss, 'ro-', label='Validation loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.xticks(epochs) 
plt.legend()
plt.show()
```

```python
results = basic_model.evaluate(x_test, y_test)
for i in range(len(basic_model.metrics_names)):
    print(basic_model.metrics_names[i]," : ", results[i])
```

The values "Test Loss" and "Test MSE" represent two different metrics used to evaluate the performance of your model on the test dataset. Here's what each one means:

Test Loss (Mean Absolute Error - MAE): This is the value you obtained as the primary loss function for your model. Since you compiled the model with loss='mean_absolute_error', the "Test Loss" refers to the Mean Absolute Error on your test dataset. The MAE is calculated as the average of the absolute differences between the predicted values and the actual values. It gives a direct interpretation of how far, on average, your predictions are from the actual values, regardless of the direction of the error. A MAE of 16.954 indicates that, on average, the predictions of the model are about 16.954 units away from the actual values.

Test MSE (Mean Squared Error): This is an additional metric you chose to monitor, as indicated by metrics=["mse"] in your model compilation step. The Mean Squared Error calculates the average of the squares of the differences between the predicted and actual values. It emphasizes larger errors more than smaller ones, as errors are squared before they are averaged. The MSE is a common metric in regression tasks because squaring the errors penalizes large deviations more severely than small ones. A MSE of 375.281 means that, on average, the square of the errors of the predictions is 375.281. MSE values are typically harder to interpret directly compared to MAE because they are not in the same unit as the original data.

In summary:

MAE (Test Loss) gives a straightforward average error magnitude.
MSE (Test MSE) gives more weight to larger errors and is in squared units of the original data.
Both are useful for evaluating the model, but they provide slightly different perspectives on its performance. Generally, you would look at both to get a comprehensive understanding of how well your model is performing.
